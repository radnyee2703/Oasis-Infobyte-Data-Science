```python
import datetime as dt
import calendar
```

```python
df1['Date'] = pd.to_datetime(df1['Date'],dayfirst=True)
df1['month_int'] = df1['Date'].dt.month
df1['month'] = df1['month_int'].apply(lambda x:calendar.month_abbr[x])
```

```python
df2['Date'] = pd.to_datetime(df2['Date'],dayfirst=True)
df2['month_int'] = df2['Date'].dt.month
df2['month'] = df2['month_int'].apply(lambda x:calendar.month_abbr[x])
```

# Data Visualisation

```python
#df1 unemployed rate state wise
df_1 = df1[['State','Estimated Unemployment Rate']].groupby('State').sum().sort_values('Estimated Unemployment Rate',ascending=False)
```

```python
#df2 unemployed rate state wise
df_2 = df2[['State','Estimated Unemployment Rate']].groupby('State').sum().sort_values('Estimated Unemployment Rate',ascending=False)
```

```python
df_1.head()
```

```python
df_2.head()
```

```python
#plot for df_1
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df_1[:10].plot(kind='bar',color='red',figsize=(30,5),ax=ax0)
ax0.set_title('Highest unemployed states')
ax0.set_xlabel('State')
ax0.set_ylabel('Number of unemployed people')
```

```python
#plot for df_2
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df_2[:10].plot(kind='bar',color='blue',figsize=(30,5),ax=ax0)
ax0.set_title('Highest unemployed states')
ax0.set_xlabel('State')
ax0.set_ylabel('Number of unemployed people')
```

```python
#df1 unemployment rate month wise
df_3 = df1[['month','Estimated Unemployment Rate']].groupby('month').sum().sort_values('Estimated Unemployment Rate',ascending=False)
df_3.head()
```

```python
#df2 unemployment rate month wise
df_4 = df2[['month','Estimated Unemployment Rate']].groupby('month').sum().sort_values('Estimated Unemployment Rate',ascending=False)
df_4.head()
```

```python
#plot for df_3
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df_3[:12].plot(kind='bar',color='yellow',figsize=(30,5),ax=ax0)
ax0.set_title('Highest unemployed states')
ax0.set_xlabel('month')
ax0.set_ylabel('Number of unemployed people')
```

```python
#plot for df_4
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df_4[:12].plot(kind='bar',color='green',figsize=(30,5),ax=ax0)
ax0.set_title('Highest unemployed states')
ax0.set_xlabel('month')
ax0.set_ylabel('Number of unemployed people')
```

```python
#df1 mean of all columns
df_5 = df1.groupby(['month'])[['Estimated Unemployment Rate','Estimated Employed', 'Estimated Labour Participation Rate']].mean()
df_5 = pd.DataFrame(df_5).reset_index()
month = df_5.month
unemp_rate = df_5['Estimated Unemployment Rate']
lpr = df_5['Estimated Labour Participation Rate']
```

```python
#plot for df_5
fig = go.Figure()
fig.add_trace(go.Bar(x=month,y=unemp_rate,name='Unemployment Rate'))
fig.add_trace(go.Bar(x=month,y=lpr,name='Labour Participation Rate'))
fig.update_layout(title='Unemployment and Labour Rate',xaxis={'categoryorder':'array','categoryarray':['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']})
fig.show()
```

```python
#df2 mean of all columns
df_6 = df2.groupby(['month'])[['Estimated Unemployment Rate','Estimated Employed', 'Estimated Labour Participation Rate']].mean()
df_6 = pd.DataFrame(df_6).reset_index()
month = df_6.month
unemp_rate = df_6['Estimated Unemployment Rate']
lpr = df_6['Estimated Labour Participation Rate']
```

```python
#plot for df_6
fig = go.Figure()
fig.add_trace(go.Bar(x=month,y=unemp_rate,name='Unemployment Rate'))
fig.add_trace(go.Bar(x=month,y=lpr,name='Labour Participation Rate'))
fig.update_layout(title='Unemployment and Labour Rate',xaxis={'categoryorder':'array','categoryarray':['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']})
fig.show()
```
